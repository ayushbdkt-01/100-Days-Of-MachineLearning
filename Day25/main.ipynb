```python
# NORMALIZATION
# technique used in data preprocessing which aims to change the values in the numeric column to use a common scale, without changing the range of information and loss of data

# main work is to eliminate the units related to numeric columns and considering only the magnitude
```

```python
# TYPES OF NORMALIZATION = minmaxscaling (used 90% times), meanNormalixation, maxAbsoluteScaling, robustScaling
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',header=None, usecols=[0,1,2])
df.columns=['ClassLabel', "Alcohol", 'Malic Acid']
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python

sns.scatterplot(x=df['Alcohol'],y=df['Malic Acid'], hue=df['ClassLabel'])
```

```python
# TRAIN_TEST_SPLIT

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(df.drop('ClassLabel',axis=1), df['ClassLabel'], test_size=0.3, random_state=0)

x_train.shape
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar=MinMaxScaler() # creating object of minMaxScalar() class

scalar.fit(x_train) # fitting scalar to training data

# transforming training and testing
x_train_scaled = scalar.transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
# scikit-learn returns a numpy array, so we converted it to DataFrame
x_train_scaled=pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
# minimum will be changed to 0
# maximum will be changed to 1
# becauses value ranges bw 0 and 1
```

```python
# Before Scaling = values in any range
# After Scaling = values are compressed between 0 and 1
```

# BiVariate Analysis and MultiVariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
flight=sns.load_dataset("flights")
iris=sns.load_dataset("iris")
titanic=pd.read_csv('train.csv')
tips=sns.load_dataset("tips")
```

```python
# scatter plot (numerical-numerical)

sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
# barplot (numerical-categorical)
# x-axis pe categorical column
# y-axis pe numerical column


# sns.barplot(x=titanic['Pclass'],y=titanic['Age'])
# sns.barplot(x=titanic['Pclass'],y=titanic['Age'], hue=titanic['Sex'])

# sns.barplot(x=titanic['Pclass'],y=titanic['Fare'])
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'], hue=titanic['Sex'])
```

```python
# box Plot (numerical-column)
# x-axis pe categorical column
# y-axis pe numerical column

sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

```python
# DISPLOT = (Numerical-Categorical)
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

```python
# heatmap (categorical-categorical)
# sns.heatmap()
pd.crosstab(titanic["Pclass"],titanic['Survived'])

sns.heatmap( pd.crosstab(titanic["Pclass"],titanic['Survived']) )

```

```python
# ClusterMap (categorical-categorical)
pd.crosstab(titanic["SibSp"],titanic['Survived'])

sns.clustermap(pd.crosstab(titanic["SibSp"],titanic['Survived']))
```

```python
# pairplot = collection of scatter plot 
sns.pairplot(iris, hue='species')
```

```python
# LINEPLOT (numerical-numerical)
# special form of scatter plot
# it is used for time based columns in x-axis

flight.head()
sns.lineplot(flight['passengers'])
```

```python
flight.pivot_table(values="passengers",index='month',columns='year')

sns.heatmap(flight.pivot_table(values="passengers",index='month',columns='year'))
```
